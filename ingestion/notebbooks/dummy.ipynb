```python
livehead_api = LiveHeatAPI()
query = """
  query getOrganisationAthletesWithMemberships($id: ID!, $page: Int!, $per: Int!) {
  organisationAthletes(id: $id, page: $page, per: $per) {
    athletes {
      id
      name
      memberships {
        id
      }
    }
    totalCount
  }
}
"""
variables = {"id": 78, "page": 1, "per": 1000}
response = livehead_api.query(query, variables)
athletes = response["organisationAthletes"]["athletes"]
total_count = response["organisationAthletes"]["totalCount"]
while len(athletes) < total_count:
    variables["page"] += 1
    response = livehead_api.query(query, variables)
    athletes += response["organisationAthletes"]["athletes"]
df = pd.DataFrame(athletes)
df.head()
```

```python
df[df['memberships'].apply(lambda x: len(x)) > 0].head()
```

```python
# normalise the federatedOrganisations
df = pd.concat([df, pd.json_normalize(df["federatedOrganisations"])], axis=1)
df = df.rename(columns={"id": "federated_organisation_id"})
# drop the federatedOrganisations column
df = df.drop(columns=["federatedOrganisations"])
df.head()
```
